```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
from pandas import read_csv, read_excel 
%matplotlib inline
```

```python
df = read_excel('response_data.xlsx')
df.shape
```

```python
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

```python
df["How satisfied are you? "].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.ylabel('')  # Hide y-label for better aesthetics
plt.title('Satisfaction Levels')
plt.show()
```

```python
df["What automation codes"] = df["What automation features  would you like? "].astype('category').cat.codes
```

```python
df.head()
```

```python
df["What automation codes"].plot.hist(bins=8, edgecolor='black')
plt.xlabel('What automation feature codes')
plt.ylabel('Frequency')
plt.title('Histogram of What automation features')
plt.show()
```

```python
# Plotting a bar chart
df['recieved_training'].value_counts().plot.bar()
plt.xlabel('Recieved Training')
plt.ylabel('Frequency')
plt.title('Bar Chart of Participants that Recieved Training')
plt.show()
```

```python
df["role"].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.ylabel('')  # Hide y-label for better aesthetics
plt.title("Participnats' role at the University lIbrary")
plt.show()
```

```python
# Plotting a count plot
sns.countplot(data=df, x='Librabry_accessibility', hue='Librabry_accessibility', native_scale= True,gap=0)
plt.xlabel('Those that recieved library Access')
plt.ylabel('Count of Participants')
plt.title('Count Plot of Participants with access to the Library')
plt.show()
```

```python
# Creating a crosstab for heatmap
plt.figure(figsize=(9,9))
heatmap_data = pd.crosstab(df['role'], df['recieved_training'])
# Plotting a heatmap
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm')
plt.title('Heatmap of individuals with Training by role')
plt.show()
```

```python
# Creating a crosstab for heatmap
heatmap_data = pd.crosstab(df['role'], df['recieved_training'])

# Plotting a heatmap
sns.scatterplot(heatmap_data)
plt.title('Heatmap of individuals with Training by role')
plt.show()
```

```python
# Plotting a pair plot
sns.pairplot(df, hue='role')
plt.show()
```
